# Descriptive Statistics

```python
# Descriptive statistics 

import numpy as np

def descriptive_statistics(sample, field):
    
    stats = {
        "Mean": sample[field].mean(),
        "Median": sample[field].median(),
        "Minimum": sample[field].min(),
        "Maximum": sample[field].max(),
        "Standard Deviation": sample[field].std()
    }
    
    return stats

whole_dataset_stats = descriptive_statistics(qol, "age_at_death")

print(whole_dataset_stats)
```

```python
#Descriptive Statistics - Measure of Central tendency

qol.median(numeric_only=True) 
qol['work_hours'].median() 
qol['age_at_death'].mode()
```

```python
#Descriptive Stistics - Measure of diapersion

qol['rest_hours'].min() 
qol['work_hours'].max()
final_range= qol['work_hours'].max() - qol['work_hours'].min() 
print(final_range) 
qol['rest_hours'].std() 
qol['rest_hours'].var() 
```

# Statistical Inference 

```python
#One Sample T-Test
from scipy.stats import ttest_1samp

hypothesized_mean = 81

t_stat, p_value = ttest_1samp(qol['age_at_death'], hypothesized_mean)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Significant difference found (Reject Null Hypothesis)")
else:
    print("Result: No significant difference found (Fail to Reject Null Hypothesis)")
```

```python
#t-test
from scipy.stats import ttest_ind

males = qol[qol['Gender'] == 'Male']['age_at_death']
females = qol[qol['Gender'] == 'Female']['age_at_death']

t_stat, p_val = ttest_ind(males, females, equal_var=False)

print(f"Mean Age (Male): {males.mean():.2f}")
print(f"Mean Age (Female): {females.mean():.2f}")
print("-" * 30)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Result: Significant difference exists (Reject Null Hypothesis)")
else:
    print("Result: No significant difference (Fail to reject Null Hypothesis)")
```

```python
# Hypothesis Testing -- Chi-Square test 

from scipy.stats import chi2_contingency

# Create the contingency table
contingency_data = pd.crosstab(qol['Occupation'], qol['age_at_death'])

# Perform the test
chi2, p, dof, expected = chi2_contingency(contingency_data)

# FIX: Make sure this is all on one line!
print('chi2=%.3f, p=%.3f' % (chi2, p))

if p > 0.05:
    print('Independent categories')
else:
    print('Dependent categories')
```

```python
# Pearson Correlation

from scipy.stats import pearsonr

corr, p_value = pearsonr(qol['work_hours'], qol['age_at_death'])

print("Correlation =", corr)
print("p-value =", p_value)
```

```python
# Spearman's Correlation

from scipy.stats import spearmanr

corr, p = spearmanr(qol["exercise_hours"], qol["age_at_death"])

print(f"Spearman's Correlation Coefficient: {corr :.4f}")
print(f"P_Value: {p_value :.4f}")
```

```python
#Check for normality

from scipy.stats import shapiro

exercise=qol['exercise_hours']
r, p = shapiro(exercise)
print('r=%.2f, p=%.30f' % (r, p))
if p > 0.05:
    print('Normal distribution')
else:
    print('Not a normal distribution')
```

```python
# identify numeric columns
numeric_cols = qol.select_dtypes(include=['number']).columns
qol[numeric_cols].corr(method='pearson')
```

```python
# Map ordinal variables to numerical ranks
from scipy.stats import spearmanr

gender_map = {'Male': 0, 'Female': 1}
qol['gender_rank'] = qol['Gender'].map(gender_map)

# Now you can use it in Spearman
r, p_value = spearmanr(qol['gender_rank'], qol['age_at_death'])
print(f"Correlation (Gender vs Age): {r:.4f}")
```
